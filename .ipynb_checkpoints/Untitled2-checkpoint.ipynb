```python
from PIL import Image
import math
import sys
import numpy as np

# For Visualization in iPython notebook
import matplotlib.pyplot as plt


def convolution(scratch,i,j):
    gaussFilter = np.array([[1, 1, 2, 2, 2, 1, 1],
              [1, 2, 2, 4, 2, 2, 1],
              [2, 2, 4, 8, 4, 2, 2],
              [2, 4, 8, 16, 8, 4, 2],
              [2, 2, 4, 8, 4, 2, 2],
              [1, 2, 2, 4, 2, 2, 1],
              [1, 1, 2, 2, 2, 1, 1]])
    sum=0
    for k in range(i-3,i+ 4):
        for l in range(j-3, j+4):
            sum += gaussFilter[k-i+3][l-j+3]*scratch[k][l]

    # Return normalized results
    return (sum/140)


def applyPrewitt(image, height, width):
    horizontal = vertical =0
    result=[]
    Gx = np.array([[-1, 0, 1],[-1, 0, 1],[-1, 0, 1]])
    Gy = np.array([[1, 1, 1],[0, 0, 0],[-1, -1, -1]])
    
    for k in range(-1, 1):
        for l in range(-1, 1):
            temp = image[horizontal + k][vertical + l]
            a = Gx[1 + l][1 + k]
            b = Gy[1 + l][1 + k]

            horizontal += abs(temp * a)
            vertical += abs(temp * b)

    result.append(horizontal)
    result.append(vertical)
    return result


# Calculate gradient magnitude and direction 
def gradient(imgHorizontal, imgVertical, i,j):          
    gradientArray =[]
    gradientArray.append(np.power(np.power(imgVertical[i][j], 2.0) + np.power(imgVertical[i][j], 2.0), 0.5))
    gradientArray.append(np.arctan2(imgVertical[i][j], imgHorizontal[i][j]))
    gradientArray.append((np.round(gradientArray[1] * (5.0 / np.pi)) + 5) % 5)          # Quantize direction
    return gradientArray


def maxSuppression(imgMagnitude, sector, r, c):
    # Suppressing pixels at the image edge
    if sector == 0:  # Horizontal
        if imgMagnitude[r, c] <= imgMagnitude[r, c - 1] or imgMagnitude[r, c] <= imgMagnitude[r, c + 1]:
            return 0
        else: return 1
    elif sector == 1:  # Northeast - SouthWest
        if imgMagnitude[r, c]!=max(imgMagnitude[r - 1, c + 1],imgMagnitude[r, c],imgMagnitude[r + 1, c - 1]):
            return 0
        else: return 1
    elif sector == 2:  # Vertical
        if imgMagnitude[r, c]!=max(imgMagnitude[r - 1, c],imgMagnitude[r, c],imgMagnitude[r + 1, c]):
            return 0
        else: return 1
    elif sector == 3:  # Northwest - Southeast
        if imgMagnitude[r, c]!=max(imgMagnitude[r - 1, c - 1],imgMagnitude[r, c],imgMagnitude[r + 1, c + 1]):
            return 0
        else: return 1

def pTileFinder(over, array, percentile):
    # percentile = 100-percentile
    array = np.array(array)                      # return pth percentile
    threshold = np.percentile(over,percentile)
    for i in range(array.shape[0]):
        for j in range(array.shape[1]):
            if(array[i][j]>=t):
                array[i][j]=255                    # Assigning value 255 to the foreground pixels
            else :
                array[i][j]=0                   # Assigning value 0 to the background pixels
    print("Threshold: ",threshold)
    return array


def nomalizedValues(arr, value):

    absoluteArr = np.array(arr)

    # Take absolute value
    for i in range(absoluteArr.shape[0]):
        for j in range(absoluteArr.shape[1]):
            absoluteArr[i][j] = abs(arr[i][j])

    maxVal = -sys.maxsize
    minVal = sys.maxsize

    for i in range(absoluteArr.shape[0]):
        for j in range(absoluteArr.shape[1]):
            maxVal = max(absoluteArr[i][j], maxVal)
            minVal = min(absoluteArr[i][j], minVal)

    result = np.array(arr)
    normalizeImage(absoluteArr, value, maxVal, minVal, result)

    return result


def normalizeImage(array, high, maxVal, minVal, result):
    for i in range(array.shape[0]):
        for j in range(array.shape[1]):
            if array[i][j] != None:
                result[i][j] = int(((array[i][j] - minVal) / (maxVal - minVal)) * high)
            else:
                result[i][j] = 0

class cannyEdgeDetector:
    img = Image.open("Images/Lena256.bmp")
    width, height = img.size
    imgCopy = np.array(img, float)

    # Gaussian Filtering
    imgArray = np.array(img)
    # scratch = np.pad(img, [3, 3], mode='constant') # to handle the pixels for which gaussian mask goes outside image border

    # gaussianImage = gaussianSmoothing(img)
    # cv2.imwrite('Outputs/filter_gauss.jpg', gaussianImage)

    for i in range(3, height - 3):
        for j in range(3, width - 3):
            # Get the dicsrete convolution for pixel[i,j]
            result = convolution(imgCopy,i,j)
            imgArray.itemset((i, j), result)

    figure, axis = plt.subplots(1, 2)
    axis[0].imshow(img,cmap='gray')
    axis[1].imshow(imgArray,cmap='gray')
    plt.show()

    # Prewitt's Operator
    imgHorizontal = imgVertical = np.array(imgArray)
    result = []

    for i in range(height):
        for j in range(width):
            result = applyPrewitt(imgArray,i,j)
            imgHorizontal.itemset((i,j),result[0])  # Storing Horizontal Gradient
            imgVertical.itemset((i,j),result[1]) # Storing Vertical Gradient

#     imgHorizontal = nomalizedValues(imgHorizontal, 255)
#     imgVertical1 = nomalizedValues(imgVertical, 255)


    figure2, axis = plt.subplots(1, 2)
    axis[0].set_title("Horizontal Gradient")
    axis[0].imshow(imgHorizontal, cmap='gray')
    axis[1].set_title("Vertical Gradient")
    axis[1].imshow(imgVertical, cmap='gray')
    plt.show()
    print("hori magnitude")
    print(imgHorizontal)
    print("verti magnitude")
    print(imgVertical)

    # Get gradient and direction

    # Intializations
    imgMagnitude, imgAngle, imgSector = np.array(imgHorizontal), np.array(imgVertical), np.array(imgVertical)
    height, width = imgHorizontal.shape[0], imgHorizontal.shape[1]
    gradientResult = []
    for i in range(height):
        for j in range(width):
            # Calculate Gradient Angle & Magnitude
            gradientResult=gradient(imgHorizontal, imgVertical, i ,j)
            # Storing Horizontal Gradient
            imgMagnitude.itemset((i,j), gradientResult[0])

            # Vertical Gradient -> Angle
            imgAngle.itemset((i,j),gradientResult[1])

            # Sector
            imgSector.itemset((i,j),gradientResult[2])

    imgMagnitude = nomalizedValues(imgMagnitude,255)

    plt.title("Gradient Magnitude")
    plt.imshow(imgMagnitude, cmap='gray')
    plt.show()
    print("Gradient magnitude")
    print(imgMagnitude)
    print(im_angle)

    # Non-maximum suppression
    gradientSuppression = np.array(imgMagnitude)
    print(gradientSuppression)
    for r in range(imgCopy.shape[0]):
        for c in range(imgCopy.shape[1]):
            # Suppress pixels
            if r == 0 or r == imgCopy.shape[0] - 1 or c == 0 or c == imgCopy.shape[1] - 1:
                gradientSuppression[r, c] = 0
                continue

            sector = im_sector[r, c] % 4
            if Nonmax_sup(imgMagnitude, sector, r, c == 0):
                gradientSuppression[r][c]=0

    print("printing gradientSuppression for the second time")
    print(gradientSuppression)
    gradientSuppression=nomalizedValues(gradientSuppression,255)
    plt.title("Non-Maxima Supression")
    plt.imshow(gradientSuppression, cmap="gray")
    plt.show()


    # Starting with P-tile Thresholding
    gradientSuppression1, gradientSuppression2, gradientSuppression3 = gradientSuppression.copy(), gradientSuppression.copy(), gradientSuppression.copy()
    pTileResult = []
    for i in range(gradientSuppression.shape[0]):
        for j in range(gradientSuppression.shape[1]):
            if gradientSuppression[i][j] > 0:
                pTileResult.append(gradientSuppression[i][j])

    print(pTileResult)
    print("gradientSuppression")
    print(gradientSuppression)
    print("gradientSuppression1")
    print(gradientSuppression1)
    print("gradientSuppression2")
    print(gradientSuppression2)
    print("gradientSuppression3")
    print(gradientSuppression3)

    pTile1 = pTileFinder(pTileResult,gradientSuppression1,10)
    plt.title("P-tile Thresholding (P=10)")
    plt.imshow(pTile1, cmap="gray")
    plt.show()

    pTile2 = pTileFinder(pTileResult,gradientSuppression2,30)
    plt.title("P-tile Thresholding (P=30)")
    plt.imshow(pTile2, cmap="gray")
    plt.show()

    pTile3 = pTileFinder(pTileResult,gradientSuppression3,50)
    plt.title("P-tile Thresholding (P=50)")
    plt.imshow(pTile3, cmap="gray")
    plt.show()
```
