```python
def gaussianSmoothing(img): 
    #image = np.pad(img, [3,3], mode='constant')
    imgArr = np.array(img) 
    gaussianArray = imgArr.copy()
    for i in np.arange(3, height-3):
        for j in np.arange(3, width-3):
            sum = applyGaussian(imgArr, i, j)
            gaussianArray[i][j] = sum
    return gaussianArray
```

```python
def applyGaussian(image, row, col):
    sum = 0
    for i in range(row-3, row+4):
        for j in range(col-3, col+4):
            sum += gaussian_mask[i-row+3][j-col+3]*image[i][j]
    return sum/140
```

```python
gaussianImage = gaussianSmoothing(img)
cv2.imwrite('Outputs/filter_gauss.jpg', img)
```
