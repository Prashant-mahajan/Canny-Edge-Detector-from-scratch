```python
from PIL import Image
from skimage.color import rgb2gray
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import cv2
```

```python
pil_img = Image.open('Lena256.bmp')
img_gray = pil_img.convert('LA')

img = np.array(list(img_gray.getdata(band=0)), float)
img.shape = (img_gray.size[1], img_gray.size[0])
plt.imshow(img, cmap='gray')
```

```python
# Making Gaussian Kernel
sigma = 1
kernel = np.zeros([7,7], dtype=int)
W = 0    # sum of elements of kenel (for normalization)
for i in range(7):
    for j in range(7):
        sq_dist = (i-4)^2 + (j-4)^2     # wrt center 
        kernel[i][j] = np.exp(-1*(sq_dist)/(2*sigma*sigma))
        W = W + kernel[i][j]
        
kernel = kernel/W

# Now we apply filter to the image
[m,n] = np.size(img_gray)
output = np.zeros([m,n], dtype=int)
image = np.pad(img_gray, [0,0], mode='constant')

print(len(image))
for i in range(1,m):
    for j in range(1,n):
        temp = image[i:i+7][i:i+7]
        temp.astype(float)
        conv = np.dot(temp*kernel)
        output[i][j] = sum(conv[:])
        
output = uint8(output)
```

```python
img_out = img.copy()

height = img.shape[0]
width = img.shape[1]

gauss = np.array([[1, 1, 2, 2, 2, 1, 1],
                 [1, 2, 2, 4, 2, 2, 1],
                 [2, 2, 4, 8, 4, 2, 2],
                 [2, 4, 8, 16, 8, 4, 2],
                 [2, 2, 4, 8, 4, 2, 2],
                 [1, 2, 2, 4, 2, 2, 1],
                 [1, 1, 2, 2, 2, 1, 1]])
sum(sum(gauss))
```

```python
for i in np.arange(2, height-2):
    for j in np.arange(2, width-2):
        sum = 0
        for k in np.arange(-2,3):
            for l in np.arange(-2, 3):
                a = img.item(i+k, j+l)
                p = gauss[2+k, 2+l]
                sum += (p*a)
        b = sum
        img_out.itemset((i,j), b)
```

```python
cv2.imwrite('filter_gauss.jpg', img_out)
cv2.imshow('image', img_out)
```
